```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Step 1: Create sample dataset
data = {
    "size": [50, 60, 70, 80, 100, 120, 150],  # house size in square meters
    "price": [150000, 180000, 210000, 240000, 300000, 360000, 450000]  # price
}

df = pd.DataFrame(data)
```

```python
df.describe()
```

```python
# Step 2: Define input and output
X = df[["size"]]   # feature
y = df["price"]    # target
```

```python
# Step 3: Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 4: Make prediction
predicted_price = model.predict(pd.DataFrame([[95]], columns=["size"]))
print(predicted_price)
```

```python
accuracy = model.score(X, y)
print(accuracy)
```

```python
# Step 5: Visualize
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("House Size")
plt.ylabel("Price")
plt.title("House Price Prediction Model")
plt.show()
```
