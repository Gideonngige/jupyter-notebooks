```python
import pandas as pd
import numpy as np
import os
import glob

# Folder containing the raw files
input_folder = "datasets/crop_prices/csv"

# Folder where cleaned files will be saved
output_folder = "datasets/crop_prices/cleaned"

os.makedirs(output_folder, exist_ok=True)

# Get all csv files
files = glob.glob(os.path.join(input_folder, "*.csv"))

for file in files:
    
    print(f"Cleaning {file}...")
    
    df = pd.read_csv(file)
    
    # Keep only needed columns
    df = df[["Commodity", "Wholesale", "Retail", "County", "Date"]]
    
    # Remove /Kg
    df["Wholesale"] = df["Wholesale"].astype(str).str.replace("/Kg", "", regex=False)
    df["Retail"] = df["Retail"].astype(str).str.replace("/Kg", "", regex=False)
    
    # Replace '-' with NaN
    df["Wholesale"] = df["Wholesale"].replace("-", np.nan)
    df["Retail"] = df["Retail"].replace("-", np.nan)
    
    # Convert to numeric
    df["Wholesale"] = pd.to_numeric(df["Wholesale"], errors="coerce")
    df["Retail"] = pd.to_numeric(df["Retail"], errors="coerce")
    
    # Fill missing values with mean
    df["Wholesale"] = df["Wholesale"].fillna(df["Wholesale"].mean())
    df["Retail"] = df["Retail"].fillna(df["Retail"].mean())
    
    # Save cleaned file
    filename = os.path.basename(file)
    output_path = os.path.join(output_folder, filename)
    
    df.to_csv(output_path, index=False)

print("✅ All crop price datasets cleaned successfully.")
```
