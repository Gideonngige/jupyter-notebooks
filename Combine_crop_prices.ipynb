```python
import pandas as pd
import glob

files = glob.glob("datasets/crop_prices/cleaned/*.csv")

price_list = []

for file in files:
    df = pd.read_csv(file)
    price_list.append(df)

prices_df = pd.concat(price_list, ignore_index=True)

prices_df
```

```python
prices_df.to_csv("datasets/raw_crop_data.csv", index=False)
```
