```python
import requests
import pandas as pd
import time

# -----------------------------------
# 1️⃣ Kenya Counties + Coordinates
# -----------------------------------
counties = {
    "Mombasa": (-4.0435, 39.6682),
    "Kwale": (-4.1737, 39.4521),
    "Kilifi": (-3.5106, 39.9093),
    "Tana River": (-1.5011, 40.0334),
    "Lamu": (-2.2697, 40.9020),
    "Taita Taveta": (-3.3980, 38.5560),
    "Garissa": (-0.4532, 39.6460),
    "Wajir": (1.7471, 40.0573),
    "Mandera": (3.9373, 41.8569),
    "Marsabit": (2.3284, 37.9899),
    "Isiolo": (0.3556, 37.5822),
    "Meru": (0.0470, 37.6498),
    "Tharaka-Nithi": (-0.2965, 37.7237),
    "Embu": (-0.5394, 37.4572),
    "Kitui": (-1.3670, 38.0106),
    "Machakos": (-1.5177, 37.2634),
    "Makueni": (-1.8041, 37.6200),
    "Nyandarua": (-0.1800, 36.5220),
    "Nyeri": (-0.4167, 36.9500),
    "Kirinyaga": (-0.6591, 37.3827),
    "Murang'a": (-0.7167, 37.1500),
    "Kiambu": (-1.1714, 36.8356),
    "Turkana": (3.1167, 35.6000),
    "West Pokot": (1.6210, 35.2300),
    "Samburu": (1.2150, 36.9540),
    "Trans Nzoia": (1.0191, 35.0023),
    "Uasin Gishu": (0.5143, 35.2698),
    "Elgeyo-Marakwet": (0.8025, 35.4780),
    "Nandi": (0.1836, 35.1260),
    "Baringo": (0.4667, 35.9667),
    "Laikipia": (0.3600, 36.7819),
    "Nakuru": (-0.3031, 36.0800),
    "Narok": (-1.0870, 35.8640),
    "Kajiado": (-1.8523, 36.7768),
    "Kericho": (-0.3670, 35.2830),
    "Bomet": (-0.7830, 35.3420),
    "Kakamega": (0.2827, 34.7519),
    "Vihiga": (0.0780, 34.7210),
    "Bungoma": (0.5690, 34.5580),
    "Busia": (0.4600, 34.1110),
    "Siaya": (0.0610, 34.2880),
    "Kisumu": (-0.1022, 34.7617),
    "Homa Bay": (-0.5273, 34.4571),
    "Migori": (-1.0634, 34.4731),
    "Kisii": (-0.6833, 34.7667),
    "Nyamira": (-0.5667, 34.9333),
    "Nairobi": (-1.2864, 36.8172)
}

# -----------------------------------
# 2️⃣ SoilGrids API Endpoint
# -----------------------------------
base_url = "https://rest.isric.org/soilgrids/v2.0/properties/query"

results = []

for county, (lat, lon) in counties.items():
    print(f"Fetching soil pH for {county}...")

    params = {
        "lat": lat,
        "lon": lon,
        "property": "phh2o",
        "depth": "0-5cm"
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    try:
        raw_ph = data["properties"]["layers"][0]["depths"][0]["values"]["mean"]
        ph_value = raw_ph / 10 if raw_ph is not None else None
    except:
        ph_value = None

    results.append({
        "County": county,
        "Latitude": lat,
        "Longitude": lon,
        "Soil_pH_0_5cm": ph_value
    })

    time.sleep(1)  # Avoid overwhelming API

# -----------------------------------
# 3️⃣ Save to CSV
# -----------------------------------
df = pd.DataFrame(results)
df.to_csv("Kenya_Counties2_Soil_pH.csv", index=False)

print("✅ Soil pH data saved as Kenya_Counties_Soil_pH.csv")
```
