```python
crop_prices_df = pd.read_csv('datasets/raw_crop_data.csv')
```

```python
crop_prices_df["County"] = crop_prices_df["County"].str.strip().str.lower()
```

```python
crop_prices_df["Wholesale"].isna().sum()
crop_prices_df["Wholesale"] = crop_prices_df["Wholesale"].fillna(crop_prices_df["Wholesale"].mean())
crop_prices_df["Wholesale"].isna().sum()
```

```python
crop_prices_df["Retail"].isna().sum()
crop_prices_df["Retail"] = crop_prices_df["Retail"].fillna(crop_prices_df["Retail"].mean())
crop_prices_df["Retail"].isna().sum()
```

```python
crop_prices_df["Crop"] = crop_prices_df["Commodity"]
```

```python
crop_prices_df = crop_prices_df[["County", "Crop", "Wholesale", "Retail","Date"]] 
```

```python
merged_df = pd.merge(
    crop_prices_df,
    climate_df,
    on=["County", "Date"],
    how="inner"
)
```

```python
# Date decomposition
import pandas as pd

# Ensure your Date column is in the correct datetime format
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Extract Year, Month, and Day
merged_df['Year'] = merged_df['Date'].dt.year
merged_df['Month'] = merged_df['Date'].dt.month
merged_df['Day'] = merged_df['Date'].dt.day

# Optional: Extract the Quarter (useful for seasonal financial trends)
merged_df['Quarter'] = merged_df['Date'].dt.quarter
```

```python
# # price spread - profitability analysis
# # Calculate the absolute difference
# merged_df['Price_Spread'] = merged_df['Retail'] - merged_df['Wholesale']

# # Calculate the Percentage Margin (often more useful for comparison across different crops)
# # Formula: (Retail - Wholesale) / Wholesale
# merged_df['Price_Margin_Pct'] = (merged_df['Price_Spread'] / merged_df['Wholesale']) * 100

# Calculate the average ratio from 'healthy' tomato rows
healthy_tomatoes = merged_df[(merged_df['Crop'] == 'Tomatoes') & (merged_df['Wholesale'] < merged_df['Retail'])]
avg_ratio = (healthy_tomatoes['Wholesale'] / healthy_tomatoes['Retail']).mean()

# Apply this ratio to the 'broken' rows
mask = (merged_df['Crop'] == 'Tomatoes') & (merged_df['Wholesale'] >= merged_df['Retail'])
merged_df.loc[mask, 'Wholesale'] = merged_df.loc[mask, 'Retail'] * avg_ratio

# Recalculate your features
merged_df['Price_Spread'] = merged_df['Retail'] - merged_df['Wholesale']
merged_df['Price_Margin_Pct'] = (merged_df['Price_Spread'] / merged_df['Wholesale']) * 100
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize encoders
county_encoder = LabelEncoder()
crop_encoder = LabelEncoder()

# Fit and transform the columns
merged_df['County_Encoded'] = county_encoder.fit_transform(merged_df['County'])
merged_df['Crop_Encoded'] = crop_encoder.fit_transform(merged_df['Crop'])

# Create mapping dictionaries for later use (Decoding)
county_mapping = dict(zip(county_encoder.classes_, county_encoder.transform(county_encoder.classes_)))
crop_mapping = dict(zip(crop_encoder.classes_, crop_encoder.transform(crop_encoder.classes_)))

print("Crop Mapping:", crop_mapping)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Sort by Date to prevent 'Data Leakage'
# (The model shouldn't see 2025 data to predict a 2024 crop)
merged_df = merged_df.sort_values('Date')

# 2. Define Features (X) and Target (y)
# We use the encoded columns we created earlier
features = ['Temp_max', 'Temp_min', 'Rainfall', 'Humidity', 'Soil_pH', 'County_Encoded', 'Month']
X = merged_df[features]
y = merged_df['Crop_Encoded']

# balance the data
from imblearn.over_sampling import SMOTE

smote = SMOTE(k_neighbors=3, random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

# 3. Perform a Chronological Split (80% Train, 20% Test)
# We set shuffle=False to keep the time order intact
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, shuffle=False
)

# 4. Initialize and Train the Random Forest
# n_estimators=100 means we are using 100 decision trees for better accuracy
suitability_model = RandomForestClassifier(n_estimators=300, max_depth=20, min_samples_leaf=2, class_weight='balanced_subsample', random_state=42)
suitability_model.fit(X_train, y_train)

# 5. Evaluate the Model
y_pred = suitability_model.predict(X_test)

print(f"--- Suitability Model Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nDetailed Performance per Crop:")
print(classification_report(y_test, y_pred))
```

```python
def recommend_crop(temp_max, temp_min, rainfall, humidity, soil_ph, county_name):
    # 1. Encode the input County
    county_enc = county_mapping[county_name]
    current_month = 5 # Example: May
    current_year = 2026

    # 2. Get Probabilities from Suitability Model
    input_data = [[temp_max, temp_min, rainfall, humidity, soil_ph, county_enc, current_month]]
    probabilities = suitability_model.predict_proba(input_data)[0]
    
    recommendations = []
    
    # 3. For each crop, predict the price and calculate a "Profit Score"
    for crop_name, crop_id in crop_mapping.items():
        prob = probabilities[crop_id]
        
        # Predict price for this specific crop
        price_input = [[crop_id, county_enc, current_month, current_year, rainfall]]
        predicted_price = price_model.predict(price_input)[0]
        
        # Profit Score = Probability of Success * Predicted Price
        profit_score = prob * predicted_price
        
        recommendations.append({
            'Crop': crop_name,
            'Suitability': f"{prob*100:.1f}%",
            'Expected_Price': round(predicted_price, 2),
            'Score': profit_score
        })

    # 4. Sort by Score (Most suitable AND profitable first)
    return sorted(recommendations, key=lambda x: x['Score'], reverse=True)

# Test it!
print(recommend_crop(19, 12, 1200, 75, 5.5, 'nakuru'))
```
