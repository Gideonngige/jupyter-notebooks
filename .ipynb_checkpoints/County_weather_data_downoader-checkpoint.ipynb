```python
import requests
import pandas as pd
from datetime import datetime

# -------------------------------
# 1. Kenya Counties + Coordinates
# -------------------------------
counties = {
    "Nairobi": {"lat": -1.286389, "lon": 36.817223},
    "Meru": {"lat": 0.047035, "lon": 37.649803},
    "Embu": {"lat": -0.539444, "lon": 37.457222},
    "Mombasa": {"lat": -4.043477, "lon": 39.668206},
    "Kisumu": {"lat": -0.091702, "lon": 34.7680},
}

# -------------------------------
# 2. Date Range
# -------------------------------
start_date = "20100101"   # Format: YYYYMMDD
end_date   = "20241231"

# -------------------------------
# 3. Parameters to Fetch
# -------------------------------
parameters = "T2M_MAX,T2M_MIN,PRECTOTCORR,RH2M"

base_url = "https://power.larc.nasa.gov/api/temporal/daily/point"

for county, coords in counties.items():
    lat = coords["lat"]
    lon = coords["lon"]

    url = (
        f"{base_url}"
        f"?parameters={parameters}"
        f"&community=AG"
        f"&longitude={lon}"
        f"&latitude={lat}"
        f"&start={start_date}"
        f"&end={end_date}"
        f"&format=JSON"
    )

    print(f"Downloading data for {county}...")

    response = requests.get(url)
    data = response.json()

    # Extract daily data
    climate_data = data["properties"]["parameter"]

    df = pd.DataFrame(climate_data)
    df.index = pd.to_datetime(df.index, format="%Y%m%d")

    # Save to CSV
    filename = f"{county}_climate_data.csv"
    df.to_csv(filename)

    print(f"Saved {filename}")

print("✅ All counties downloaded successfully!")
```
