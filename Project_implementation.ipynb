```python
import pandas as pd
import numpy as np
```

```python
climate_df = pd.read_csv('datasets/weather/Kenya_All_Counties_Climate_Data_2010_2024.csv')
soil_df = pd.read_csv('datasets/soil/Kenya_Counties_Soil_pH.csv')
```

```python
# clean county names
climate_df["County"] = climate_df["County"].str.strip().str.lower()
soil_df["County"] = soil_df["County"].str.strip().str.lower()
```

```python
soil_df
```

```python
# filling null value for mombasa by using near couties mean
# Define neighbors
mombasa_neighbors = ["kwale", "kilifi", "taita taveta"]

# Calculate average soil pH
mombasa_avg_ph = soil_df[
    soil_df["County"].isin(mombasa_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Mombasa soil pH:", mombasa_avg_ph)

# filling null values for nairobi
nairobi_neighbors = ["kiambu", "machakos", "kajiado", "murang'a"]

nairobi_avg_ph = soil_df[
    soil_df["County"].isin(nairobi_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Nairobi soil pH:", nairobi_avg_ph)

# filling null values for garissa
garissa_neighbors = ["wajir", "marsabit"]

garissa_avg_ph = soil_df[
    soil_df["County"].isin(garissa_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Garissa soil pH:", garissa_avg_ph)

# filling null values for trans nzoia
trans_neighbors = ["turkana", "west pokot", "samburu", "uasin gishu"]

trans_avg_ph = soil_df[
    soil_df["County"].isin(trans_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Trans Nzoia soil pH:", trans_avg_ph)

# filling null values for nakuru
nakuru_neighbors = ["laikipia", "narok", "kajiado"]

nakuru_avg_ph = soil_df[
    soil_df["County"].isin(nakuru_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Nakuru soil pH:", nakuru_avg_ph)

# filling null values for nakuru
kakamega_neighbors = ["vihiga", "siaya", "kisii", "nyamira"]

kakamega_avg_ph = soil_df[
    soil_df["County"].isin(kakamega_neighbors)
]["Soil_pH_0_5cm"].mean()

print("Estimated Kakamega soil pH:", kakamega_avg_ph)
```

```python
# filling the missing values
# Fill Mombasa
soil_df.loc[
    (soil_df["County"] == "mombasa") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = mombasa_avg_ph

# fill for lamu use for momabase bcz they are on the same region
soil_df.loc[
    (soil_df["County"] == "lamu") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = mombasa_avg_ph

# Fill Nairobi
soil_df.loc[
    (soil_df["County"] == "nairobi") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = nairobi_avg_ph


# Fill Garissa
soil_df.loc[
    (soil_df["County"] == "garissa") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = garissa_avg_ph

# Fill Mandera use Garissa because they are on the same region
soil_df.loc[
    (soil_df["County"] == "mandera") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = garissa_avg_ph

# Fill Trans Nzoia
soil_df.loc[
    (soil_df["County"] == "trans nzoia") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = trans_avg_ph

# Fill Nakuru
soil_df.loc[
    (soil_df["County"] == "nakuru") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = nakuru_avg_ph

# Fill Kericho - use nakuru
soil_df.loc[
    (soil_df["County"] == "kericho") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = nakuru_avg_ph

# Fill bomet - use nakuru
soil_df.loc[
    (soil_df["County"] == "bomet") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = nakuru_avg_ph

# Fill kakamega
soil_df.loc[
    (soil_df["County"] == "kakamega") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph

# Fill bungoma
soil_df.loc[
    (soil_df["County"] == "bungoma") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph

# Fill busia - use kakamega
soil_df.loc[
    (soil_df["County"] == "busia") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph

# Fill kisumu - use kakamega
soil_df.loc[
    (soil_df["County"] == "kisumu") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph

# Fill homa bay - use kakamega
soil_df.loc[
    (soil_df["County"] == "homa bay") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph

# Fill migori - use kakamega
soil_df.loc[
    (soil_df["County"] == "migori") &
    (soil_df["Soil_pH_0_5cm"].isnull()),
    "Soil_pH_0_5cm"
] = kakamega_avg_ph
```

```python
import pandas as pd
import numpy as np
import os
import glob

# Folder containing the raw files
input_folder = "datasets/crop_prices/csv"

# Folder where cleaned files will be saved
output_folder = "datasets/crop_prices/cleaned"

os.makedirs(output_folder, exist_ok=True)

# Get all csv files
files = glob.glob(os.path.join(input_folder, "*.csv"))

for file in files:
    
    print(f"Cleaning {file}...")
    
    df = pd.read_csv(file)
    
    # Keep only needed columns
    df = df[["Commodity", "Wholesale", "Retail", "County", "Date"]]
    
    # Remove /Kg
    df["Wholesale"] = df["Wholesale"].astype(str).str.replace("/Kg", "", regex=False)
    df["Retail"] = df["Retail"].astype(str).str.replace("/Kg", "", regex=False)
    
    # Replace '-' with NaN
    df["Wholesale"] = df["Wholesale"].replace("-", np.nan)
    df["Retail"] = df["Retail"].replace("-", np.nan)
    
    # Convert to numeric
    df["Wholesale"] = pd.to_numeric(df["Wholesale"], errors="coerce")
    df["Retail"] = pd.to_numeric(df["Retail"], errors="coerce")
    
    # Fill missing values with mean
    df["Wholesale"] = df["Wholesale"].fillna(df["Wholesale"].mean())
    df["Retail"] = df["Retail"].fillna(df["Retail"].mean())
    
    # Save cleaned file
    filename = os.path.basename(file)
    output_path = os.path.join(output_folder, filename)
    
    df.to_csv(output_path, index=False)

print("✅ All crop price datasets cleaned successfully.")
```

```python
import pandas as pd
import glob

files = glob.glob("datasets/crop_prices/cleaned/*.csv")

price_list = []

for file in files:
    df = pd.read_csv(file)
    price_list.append(df)

prices_df = pd.concat(price_list, ignore_index=True)

prices_df
```

```python
prices_df["Date"] = pd.to_datetime(prices_df["Date"])
prices_df["Year"] = prices_df["Date"].dt.year

price_summary = prices_df.groupby(["County","Year","Commodity"]).agg({
    "Wholesale":"mean",
    "Retail":"mean"
}).reset_index()

price_summary
```

```python
# print(merged_summary["County"].unique())
# print(price_summary["County"].unique())
def clean_county(name):
    return (
        name.lower()
        .replace("-", " ")
        .replace("'", "")
        .strip()
    )

merged_summary["County"] = merged_summary["County"].apply(clean_county)
price_summary["County"] = price_summary["County"].apply(clean_county)
print(merged_summary["County"].unique())
print(price_summary["County"].unique())
```

```python
# merge climate and price datasets
merged_df = pd.merge(
    merged_summary,
    price_summary,
    on=["County","Year"],
    how="inner"
)

merged_df
```

```python
merged_df["AvgTemp"] = (merged_df["T2M_MAX"] + merged_df["T2M_MIN"]) / 2
merged_df["Price"] = (merged_df["Wholesale"] + merged_df["Retail"]) / 2
merged_df["Soil_pH"] = merged_df["Soil_pH_0_5cm"]
merged_df["Humidity"] = merged_df["RH2M"]
merged_df["Rainfall"] = merged_df["PRECTOTCORR"]
merged_df["Crop"] = merged_df["Commodity"]
# merged_df
merged_df["Price"] = merged_df["Price"].fillna(
    merged_df.groupby("Crop")["Price"].transform("mean")
)

ml_df = merged_df[[
    "County",
    "Crop",
    "AvgTemp",
    "Rainfall",
    "Humidity",
    "Soil_pH",
    "Price",
]]
ml_df["Price"].isna().sum()
```
