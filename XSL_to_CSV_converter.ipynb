```python
import pandas as pd
import glob
import os

# 1. Set your source and destination folders
source_folder = "datasets/crop_prices/"
output_folder = "datasets/crop_prices/csv/"

# Ensure the output directory exists
os.makedirs(output_folder, exist_ok=True)

# 2. Find all .xls files in the source folder
# Use *.xls for older Excel files or *.xlsx for newer ones
files = glob.glob(os.path.join(source_folder, "*.xls"))

print(f"Found {len(files)} files to convert...")

# 3. Loop through and convert
for file_path in files:
    # Get the filename without the path and extension (e.g., 'coffee_market_prices')
    file_name = os.path.basename(file_path).replace(".xls", "")
    
    try:
        # Load the file
        df = pd.read_excel(file_path)
        
        # Define the new CSV path
        csv_path = os.path.join(output_folder, f"{file_name}.csv")
        
        # Save as CSV
        df.to_csv(csv_path, index=False)
        print(f"Converted: {file_name}.xls -> {file_name}.csv")
        
    except Exception as e:
        print(f"Error converting {file_name}: {e}")

print("\nAll conversions complete!")
```
